### Chronbach's Alpha

```python
# Creating copies of the dataframes for Cronbach's Alpha computation
data_treatment_A_alpha = data_treatment_A.iloc[2:].copy()
data_treatment_B_alpha = data_treatment_B.iloc[2:].copy()
```

```python
# Converting the relevant columns to numeric
data_treatment_A_alpha = data_treatment_A_alpha.apply(pd.to_numeric, errors='coerce')
data_treatment_B_alpha = data_treatment_B_alpha.apply(pd.to_numeric, errors='coerce')
```

```python
# Making a function to calculate cronbachs_alpha #ChatGPT
def cronbach_alpha(df):
    k = df.shape[1]
    y = df.sum(axis=1)
    total_var = y.var(ddof=1)
    variances = df.var(axis=0, ddof=1)
    sum_var = variances.sum()
    alpha = (k / (k - 1)) * (1 - (sum_var / total_var))
    return alpha
```

```python
# Perceived usefulness for treatment A and B
usefulness_cols_A = [f"Q{i}A_num_use" for i in range(10, 16)]
usefulness_cols_B = [f"Q{i}B_num_use" for i in range(10, 16)]

alpha_usefulness_A = cronbach_alpha(data_treatment_A_alpha[usefulness_cols_A])
alpha_usefulness_B = cronbach_alpha(data_treatment_B_alpha[usefulness_cols_B])

print(f"Cronbach's Alpha for Perceived Usefulness (Treatment A): {alpha_usefulness_A}")
print(f"Cronbach's Alpha for Perceived Usefulness (Treatment B): {alpha_usefulness_B}")

# Privacy concern for treatment A and B
privacy_cols_A = [f"Q{i}A_num_priv" for i in range(16, 21)]
privacy_cols_B = [f"Q{i}B_num_priv" for i in range(16, 21)]

alpha_privacy_A = cronbach_alpha(data_treatment_A_alpha[privacy_cols_A])
alpha_privacy_B = cronbach_alpha(data_treatment_B_alpha[privacy_cols_B])

print(f"Cronbach's Alpha for Privacy Concern (Treatment A): {alpha_privacy_A}")
print(f"Cronbach's Alpha for Privacy Concern (Treatment B): {alpha_privacy_B}")

# Inclination towards personalization
inclination_cols = [f"Q{i}_num_gen" for i in range(6, 10)]

alpha_inclination_A = cronbach_alpha(data_treatment_A_alpha[inclination_cols])
alpha_inclination_B = cronbach_alpha(data_treatment_B_alpha[inclination_cols])

print(f"Cronbach's Alpha for Inclination Towards Personalization (Treatment A): {alpha_inclination_A}")
print(f"Cronbach's Alpha for Inclination Towards Personalization (Treatment B): {alpha_inclination_B}")
```

```python
# Defining function that groups the age differently
def regroup_age(age):
    if age in [1, 2]:
        return 'Gen Z'
    elif age in [3, 4]:
        return 'Millennials & Early Gen X'
    else:  # Combining age groups [5, 6, 7]
        return 'Late Gen X, Baby Boomers & Older'

# Apply the adjusted function to the age column for both treatments
data_treatment_A['Generation'] = data_treatment_A['Q1_num_demo'].apply(regroup_age)
data_treatment_B['Generation'] = data_treatment_B['Q1_num_demo'].apply(regroup_age)

# Checking the distribution for Treatment A
print("Age distribution for Treatment A:")
print(data_treatment_A['Generation'].value_counts())

# Checking the distribution for Treatment B
print("\nAge distribution for Treatment B:")
print(data_treatment_B['Generation'].value_counts())
```

 ### <b>ANOVA (Analysis of Variance)</b> is a statistical method used to test differences between two or more means. In this case, we want to see if there's a difference in perceived usefulness across different age groups. Each age group represents a different "mean" or average score of perceived usefulness.

 ### If age had no effect on perceived usefulness, we'd expect all age groups to have roughly the same average score. If age does have an effect, some age groups will have significantly higher or lower scores than others.

```python
# Calculating AVG perceived usefulness for each respondent for Treatment A
usefulness_cols_A = ['Q10A_num_use', 'Q11A_num_use', 'Q12A_num_use', 'Q13A_num_use', 'Q14A_num_use', 'Q15A_num_use']
data_treatment_A['avg_usefulness_A'] = data_treatment_A[usefulness_cols_A].mean(axis=1)

# ANOVA for Treatment A
fvalue_A, pvalue_A = stats.f_oneway(
    data_treatment_A['avg_usefulness_A'][data_treatment_A['Generation'] == 'Gen Z'],
    data_treatment_A['avg_usefulness_A'][data_treatment_A['Generation'] == 'Millennials & Early Gen X'],
    data_treatment_A['avg_usefulness_A'][data_treatment_A['Generation'] == 'Late Gen X, Baby Boomers & Older']
)
print(f"ANOVA result for Treatment A: F={fvalue_A}, p={pvalue_A}")
```

```python
# Calculate average perceived usefulness for each respondent for Treatment B
usefulness_cols_B = [col.replace('A', 'B') for col in usefulness_cols_A]
data_treatment_B['avg_usefulness_B'] = data_treatment_B[usefulness_cols_B].mean(axis=1)

# ANOVA for Treatment B
fvalue_B, pvalue_B = stats.f_oneway(
    data_treatment_B['avg_usefulness_B'][data_treatment_B['Generation'] == 'Gen Z'],
    data_treatment_B['avg_usefulness_B'][data_treatment_B['Generation'] == 'Millennials & Early Gen X'],
    data_treatment_B['avg_usefulness_B'][data_treatment_B['Generation'] == 'Late Gen X, Baby Boomers & Older']
)

print(f"ANOVA result for Treatment B: F={fvalue_B}, p={pvalue_B}")
```

For the second treatment (B) The p-value is <b>0.02537</b>, which is less than <b>0.05</b>. This suggests that there is a statistically significant difference in perceived usefulness across the age groups for Treatment B. Given this result, we will to conduct post-hoc tests to determine which specific age groups differ from each other in their perceptions.

```python
# Conducting Tukey's HSD post-hoc test for Treatment B
posthoc_B = pairwise_tukeyhsd(data_treatment_B['avg_usefulness_B'], data_treatment_B['Generation'])
print("Post-hoc results for Treatment B:")
print(posthoc_B)
```

Given the Tukey HSD results for Treatment B:

The post-hoc analysis using the Tukey HSD test revealed a statistically significant difference in perceived usefulness between the "Gen Z" group and the "Millennials & Early Gen X" group (p=0.0349). Specifically, the "Millennials & Early Gen X" group perceived the interface as more useful by an average difference of 1.4773 units compared to the "Gen Z" group.

However, there were no statistically significant differences in perceived usefulness between the "Gen Z" group and the "Late Gen X, Baby Boomers & Older" group (p=0.07), even though the p-value is close to the 0.05 threshold. Similarly, the difference between the "Late Gen X, Baby Boomers & Older" group and the "Millennials & Early Gen X" group was not statistically significant (p=0.8997).

In conclusion, The most notable finding is the significant difference between the "Gen Z" group and the "Millennials & Early Gen X" group. Specifically, the "Millennials & Early Gen X" group perceived the interface as more useful than the "Gen Z" group. This suggests that the younger generation (Gen Z) might have different expectations or criteria for what they consider a useful personalized ad interface compared to the slightly older generation (Millennials & Early Gen X).

## <b>H2a: Consumers exposed to the biometric scanning camera (explicit data collection) will exhibit higher privacy concerns than those exposed to the QR code scanner (implicit data collection).</b>

```python
# Extracting relevant columns for privacy concerns
privacy_cols_A = ['Q16A_num_priv', 'Q17A_num_priv', 'Q18A_num_priv', 'Q19A_num_priv', 'Q20A_num_priv']
privacy_cols_B = [col.replace('A', 'B') for col in privacy_cols_A]
```

```python
# Calculate average privacy concern for each respondent
data_treatment_A['avg_privacy_concern_A'] = data_treatment_A[privacy_cols_A].mean(axis=1)
data_treatment_B['avg_privacy_concern_B'] = data_treatment_B[privacy_cols_B].mean(axis=1)
```

```python
from scipy.stats import levene

# Levene's test for equality of variances
stat, p_val_levene = levene(data_treatment_A['avg_privacy_concern_A'].dropna(), 
                            data_treatment_B['avg_privacy_concern_B'].dropna())

print(f"Levene's test p-value: {p_val_levene}")
```

```python
from scipy.stats import ttest_ind
```

```python
# Conducting the appropriate t-test
if p_val_levene > 0.05:
    # If p-value is greater than 0.05, variances can be assumed to be equal
    t_stat, p_val = ttest_ind(data_treatment_A['avg_privacy_concern_A'].dropna(), 
                              data_treatment_B['avg_privacy_concern_B'].dropna(), 
                              equal_var=True)
else:
    # If p-value is less than 0.05, use Welch's t-test (equal_var=False)
    t_stat, p_val = ttest_ind(data_treatment_A['avg_privacy_concern_A'].dropna(), 
                              data_treatment_B['avg_privacy_concern_B'].dropna(), 
                              equal_var=False)

print(f"\nt-statistic: {t_stat}")
print(f"p-value: {p_val}")
```

### Results:
In examining the potential difference in privacy concerns between consumers exposed to the biometric scanning camera (explicit data collection) and those exposed to the QR code scanner (implicit data collection), we first tested the assumption of homogeneity of variances using Levene's test. The result of Levene's test (p = 0.706) indicated that the variances of privacy concerns between the two groups were not significantly different, allowing us to proceed with the assumption of equal variances.

Subsequently, an independent samples t-test was conducted to compare the average privacy concerns between the two treatments. The results revealed no statistically significant difference in privacy concerns between consumers exposed to the biometric scanning camera and those exposed to the QR code scanner, t(78) = -1.2405, p = 0.2185.

In conclusion, our findings do not support the hypothesis H2a. The type of data collection method, whether explicit (biometric scanning camera) or implicit (QR code scanner), did not significantly influence consumers' privacy concerns in this study.
